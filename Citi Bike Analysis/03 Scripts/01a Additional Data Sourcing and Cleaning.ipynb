# Data Cleaning

```python
# convert start and end time columns from object to timestamp
data['start_time'] = pd.to_datetime(data['start_time'])
data['end_time'] = pd.to_datetime(data['end_time'])
```

```python
# convert columns to categories
cols = ['start_station_name', 'end_station_name', 'bike_id', 'subscriber', 'gender']
for col in cols:
    data[col] = data[col].astype('category')
```

```python
data.info()
```

# Add new columns

```python
# create new column for age of customer
data['birth_year'] = pd.to_datetime(data['birth_year'])
data['age'] = (data['start_time'] - data['birth_year']) / np.timedelta64(1, 'Y')
```

```python
# create trip_minutes column to convert trip time from seconds to minutes
data['trip_minutes'] = data['trip_duration'] // 60
```
