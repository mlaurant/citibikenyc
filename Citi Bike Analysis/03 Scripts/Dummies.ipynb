```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
```

```python
# create path
path = r'/Users/mimielaurant/Documents/Career Foundry/Immersion_Achievement 6/Citibike Project'
```

```python
data = pd.read_csv(os.path.join(path, '02 Data','Prepared Data', 'citibikes_fresh.csv'), index_col = False)
```

```python
data.head()
```

```python
ride_counts = pd.get_dummies(data, columns = ['weekday'], drop_first = True)
```

```python
ride_counts
```

```python
ride_counts.columns
```

```python
# export 
ride_counts.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'citibikes_dummies.csv'))
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# Plotting each dummy column
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Mon'], label='Monday', color='midnightblue')
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Wed'], label='Wednesday', color='navy')
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Thu'], label='Thursday', color='orangered')
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Fri'], label='Friday', color='mediumblue')
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Sat'], label='Saturday', color='orange')
plt.scatter(ride_counts['start_hour'], ride_counts['weekday_Sun'], label='Sunday', color='blue')

# Adding labels and title
plt.xlabel('Start Hour')
plt.ylabel('Dummy Values')
plt.title('Scatter Plot of Start Hour and Dummy Columns')
plt.legend()
```
