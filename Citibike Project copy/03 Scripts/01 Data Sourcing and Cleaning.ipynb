# Consistency Checks

```python
# check for mixed-type columns

for col in df.columns.tolist():
  weird = (df [[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# check for missing values

df.isnull().sum()
```

```python
df.median()
```

```python
# create subset of data with missing birth years

df_nan = df[df['birth_year'].isnull() == True]
```

```python
df_nan
```

```python
# impute median value into missing birth years 

df['birth_year'] = df['birth_year'].fillna(df['birth_year'].median())
```

I chose to impute the median instead of mean to try to account for any outliers, and to reduce the risk of skewing the data. I did not want to delete 13% of the data and risk further analysis being drastically altered.

```python
# second check for missing values

df.isnull().sum()
```

```python
# check for duplicates
df_dups = df[df.duplicated()]
```

In the descriptive stats above, it says that the minimum birth year is 1899, I highly doubt that someone born in 1899 woul dbe renting a bike in 2013. I'm going to check the birth year colun for extreme values.

```python
# check for birth years prior to 1913

df[df['birth_year'] <1913]
```

Since there are less than 30 records with birth years prior to 1913, I will delete them from the dataset. Choosing 1913 was generous, however there are some spry centenarians out there!

```python
df_new = df[df['birth_year'] > 1913]
```
