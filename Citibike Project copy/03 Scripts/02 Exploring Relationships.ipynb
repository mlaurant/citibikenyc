# Correlations

```python
# create subset for inital correlation testing
data_2 = data[['start_hour', 'start_station_id', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_latitude', 'end_station_longitude','trip_duration', 'gender']]
```

```python
data_2.head()
```

```python
# create a correlation matrix using pandas
data_2.corr()
```

```python
# create a correlation heatmap using matplotlib
plt.matshow(data_2.corr())
plt.show()
```

```python
# export to png file
plt.matshow(data_2.corr())
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'rough_heatmap.png'))
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(data_2.corr(), fignum=f.number) # type of plot
plt.xticks(range(data_2.shape[1]), data_2.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(data_2.shape[1]), data_2.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(data_2.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

The strongest relationship (0.59) exists between the start_station_latitude and end_station_latitude. There is also a moderate relationship (0.42) between the start_station_longitude and the end_station_longitude. I can interpret this in two ways, (1) most customers are travelling east to west and vice versa as opposed to north to south and vice versa, or (2) the docking stations are positioned on streets that run east-west across the city.

# Categorical Plots

```python
# create a histogram of trip duration
sns.histplot(data['trip_duration'], bins = 30, kde = True)
```

```python
data.loc[data['trip_duration'] < 300, 'Trip Range'] = 'short trip'
```

```python
data.loc[(data['trip_duration'] >= 300) & (data['trip_duration'] < 900), 'Trip Range'] = 'average trip'
```

```python
data.loc[data['trip_duration'] >= 900, 'Trip Range'] = 'long trip'
```

```python
data['Trip Range'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above
sns.set(style="ticks")
catplot = sns.catplot(x="start_station_latitude", y="end_station_latitude", hue="Trip Range", data=data)
```

Based on this visual, it appears as if there is a strong, postitive correlation between short trips and the latitude of the starting and ending stations. The average and long strips also appear to have a positive correlation, but not as strong as the one for short trips.
