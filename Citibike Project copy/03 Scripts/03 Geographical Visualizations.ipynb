# Wrangling Data

```python
# create locationslist dataframe for map markers
locations = data[['start_station_latitude', 'start_station_longitude']]
```

```python
# locate duplicates
loc_dups = locations[locations.duplicated()]
```

```python
# remove duplicates from locations dataframe
locations_clean = locations.drop_duplicates()
```

```python
locations_clean.shape
```

```python
# create locations list
locationlist = locations_clean.values.tolist()
len(locationlist)
locationlist[7]
```

# Generating a map using Folium

```python
map = folium.Map(location=[40.712776, -74.005974], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=data['start_station_name'][point]).add_to(map)
map
```
